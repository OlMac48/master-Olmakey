### Построим модели

#### Модель первого региона

```python
def prep_data(df):
# разделим  датасет на тренировочную и валидационную выборки

    X_train, X_valid, y_train, y_valid = train_test_split(
                                                          df.drop(['id', 'product'], axis=1),
                                                          df['product'],
                                                          random_state=RANDOM_STATE,
                                                          test_size=TEST_SIZE)
#инициализируем скелер
    scaler = StandardScaler()   
#инициализируем модель 
    model = LinearRegression()    
#преобразовываем количественные данные
    X_train_scaled = scaler.fit_transform(X_train)
    X_valid_scaled = scaler.transform(X_valid)
#обучаем модель на тренировочной выборке
    model.fit(X_train_scaled, y_train)
#делаем предсказание на валидационной выборке
    preds= model.predict(X_valid_scaled) 
# получение значения RMSE
    rmse = mean_squared_error(y_valid, preds, squared=False).round(2)
# добавим предсказанные значения в датафрейм для валидационной выборки
# выделим индексы валидационной выборки
    valid_index = X_valid.index
#отберем данные из исходного файла по индексам валидационной выборки
    df_valid_preds = df.loc[valid_index, :]
#добавим столбец предсказаний
    df_valid_preds['preds'] = preds
    return preds, rmse, df_valid_preds
```

```python
#вариант функции с полиномизацией признаков
def prep_data_poly(df, deg=1):
# разделим  датасет на тренировочную и валидационную выборки

    X_train, X_valid, y_train, y_valid = train_test_split(
                                                          df.drop(['id', 'product'], axis=1),
                                                          df['product'],
                                                          random_state=RANDOM_STATE,
                                                          test_size=TEST_SIZE)
#инициализируем скелер
    scaler = StandardScaler()   
#инициализируем модель 
    model = LinearRegression()    
#преобразовываем количественные данные
    X_train_scaled = scaler.fit_transform(X_train)
    X_valid_scaled = scaler.transform(X_valid)
    
# создаём и обучаем экземпляр объекта для полиномизации
    if deg == 1:
        X_poly_train = X_train_scaled
        X_poly_valid = X_valid_scaled
    else :
        poly = PolynomialFeatures(degree=deg)
        X_poly_train = poly.fit_transform(X_train_scaled)
        X_poly_valid = poly.transform(X_valid_scaled)
        poly.fit(X_poly_train, y_train)

#обучаем модель на тренировочной выборке
    model.fit(X_poly_train, y_train)
#делаем предсказание на валидационной выборке
    preds= model.predict(X_poly_valid) 
# получение значения RMSE
    rmse = mean_squared_error(y_valid, preds, squared=False).round(2)
# добавим предсказанные значения в датафрейм для валидационной выборки
# выделим индексы валидационной выборки
    valid_index = X_valid.index
#отберем данные из исходного файла по индексам валидационной выборки
    df_valid_preds = df.loc[valid_index, :]
#добавим столбец предсказаний
    df_valid_preds['preds'] = preds
    return preds, rmse, df_valid_preds
```

```python
preds_geo_data_0, rmse_geo_data_0, df_0_valid_preds = prep_data(df_geo_data_0)
print('Величина среднеквадратичной ошибки модели первого региона', rmse_geo_data_0)
print('Среднее значение объема нефти первого региона:', df_geo_data_0['product'].mean().round(2))
print('Среднее предсказанное значение объема нефти первого региона:', df_0_valid_preds['preds'].mean().round(2))
display('Валидационная выборка первого региона с предсказаниями', df_0_valid_preds.head(10))
```

Предсказанное значение оцень близко фактическому среднему значению объема нефти в скважине в первом регионе, но RMSE имеет высокое значение - 37,69.
Попробуем использовать полиномизацию признаков.

```python
#займемся настройкой гиперпараметра - один гиперпараметр — степень в методе PolynomialFeatures.
for num in range(1, 10):
    preds_geo_data_0, rmse_geo_data_0, df_0_valid_preds = prep_data_poly(df_geo_data_0, num)
    print(f'Для степени {num} величина среднеквадратичной ошибки модели первого региона', rmse_geo_data_0,'\n',\
            'среднее значение объема нефти первого региона:', df_geo_data_0['product'].mean().round(2),'\n',\
            'среднее предсказанное значение объема нефти первого региона:', df_0_valid_preds['preds'].mean().round(2),'\n')
```

На платформе расчет зависал - доходил до степени 6 и далеее уходил в зависание. На двух рабочих машинах и ноутбуке отлично отрабатывал. Привожу результат здесь.

Без него рассматривать далее не стоит - действительно результаты улучшились.

На первый взгляд полиномизация не оказала существенного вляния на качество модели, но оказалось, что результаты улучшились и поэтому не хотельлось бы от этого этапа отказываться из-за торможений платформы.

## Подготовительный этап расчета прибыли

По условиям задачи 

 - При разведке региона исследуют 500 точек, из которых с помощью машинного обучения выбирают 200 лучших для разработки.
 - Бюджет на разработку скважин в регионе — 10 млрд рублей.
 - При нынешних ценах один баррель сырья приносит 450 рублей дохода. Доход с каждой единицы продукта составляет 450 тыс. рублей, поскольку объём указан в тысячах баррелей.
 - После оценки рисков нужно оставить лишь те регионы, в которых вероятность убытков меньше 2.5%. Среди них выбирают регион с наибольшей средней прибылью.


### Ключевые значения для расчетов

```python
well_col = 200
budget_well = 10000000000
price_barrel = 450
price = 450000
```

```python
#функция выборки 200 лучших
def volume_geo(df_geo, ncol='preds'):
    vol = df_geo.sort_values(by=ncol, ascending=False)[0:200]['preds'].mean().round(2)
    return vol
```

```python
# рассчитаем допустимый минимальный средний объем нефти скважины
volume_well = budget_well/well_col/price
print(f'Минимально-\033[1mдопустимый средний объем\033[0m нефти скважины: \033[1m{volume_well:0.4}\033[0m тысяч баррелей')
# оценим объемы 200 лучших скважин в трех регионах
print(f'Среднее значение объема лидеров \033[1mпервого\033[0m региона:\
 \033[1m{volume_geo(df_0_valid_preds)}\033[0m тысяч баррелей;\n \
                               \033[1mвторого\033[0m региона: \033[1m{volume_geo(df_1_valid_preds)}\033[0m тысяч баррелей;\n \
                              \033[1mтретьего\033[0m региона: \033[1m{volume_geo(df_2_valid_preds)}\033[0m тысяч баррелей.')
```

**Вывод**

Данные подготовлены. 200 лучших скважин во всех регионах имеют запас больше минимально-допустимого.

## Расчет прибыли скважин

```python
#функция расчета прибыли - отбор лучших, сумма целевых объемов на стоимость тонны минус затраты на освоение
def profit_well(df_geo):
    df_geo_ = df_geo.sort_values(by='preds', ascending=False)[:well_col]
    return (df_geo_['product'].sum()*price - budget_well)   
```

```python
def format_number(x):
        return '{:,.3f}'.format(x).replace(',', ' ')
```

```python
print(f'Ожидаемая прибыль \033[1mпервого\033[0m региона:\
 \033[1m{format_number(profit_well(df_0_valid_preds))}\033[0m рублей;\n \
                 \033[1mвторого\033[0m региона: \033[1m{format_number(profit_well(df_1_valid_preds))}\033[0m рублей;\n \
                \033[1mтретьего\033[0m региона: \033[1m{format_number(profit_well(df_2_valid_preds))}\033[0m рублей.')
```

## Расчет рисков

### Распределение прибыли

По условиям задачи необходимо найти распределение прибыли (применив технику Bootstrap с 1000 выборок).

 Найдем среднюю прибыль, 95%-й доверительный интервал и риск убытков(Убыток — это отрицательная прибыль).

Также 

•	После оценки рисков нужно оставить лишь те регионы, в которых вероятность убытков меньше 2.5%. Среди них выбирают регион с наибольшей средней прибылью.

```python
#создадим функцию для бутстрепа

state = np.random.RandomState(12345)

def profit_geo(df_geo):
    values = []
    for i in range (1000):
        df_geo_sample = df_geo.sample(500, replace=True, random_state = state)
        values.append(profit_well(df_geo_sample))
    values=pd.Series(values)
    return values
```

Вероятно, по условию задачи необходимо определить вероятность убытка (риск). Поэтому найдем с помощью бутстрапа возможную прибыль и границы 95% интервала -  сделаем для трех регионов.

### Средняя прибыль, убыток

```python
#рассчитаем предполагаемую прибыль с помощью бутстрепа
df_0_profit = profit_geo(df_0_valid_preds)
df_1_profit = profit_geo(df_1_valid_preds)
df_2_profit = profit_geo(df_2_valid_preds)
```

```python
print(f'Средняя прибыль \033[1mпервого\033[0m региона:\
 \033[1m{format_number(df_0_profit.mean())}\033[0m рублей;\n \
               \033[1mвторого\033[0m региона: \033[1m{format_number(df_1_profit.mean())}\033[0m рублей;\n \
              \033[1mтретьего\033[0m региона: \033[1m{format_number(df_2_profit.mean())}\033[0m рублей.')
```
